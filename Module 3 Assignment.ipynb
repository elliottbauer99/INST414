```python
# Step 1: Define a function to calculate Euclidean distance with widened SAT range
def euclidean_distance(college1, college2, sat_range=200):
    features = ['LATITUDE', 'LONGITUDE']
    if 'SAT_AVG' in college1.index and 'SAT_AVG' in college2.index:
        features.append('SAT_AVG')
    dist_lat_long = np.linalg.norm([college1[feat] - college2[feat] for feat in features[:-1]])
    if 'SAT_AVG' in features:
        if abs(college1['SAT_AVG'] - college2['SAT_AVG']) <= sat_range:
            dist_sat = 0
        else:
            dist_sat = 1
        return np.sqrt(dist_lat_long**2 + dist_sat**2)
    else:
        return dist_lat_long

# Step 2: Define a function to find most similar schools with widened SAT range
def find_similar_schools(user_input, data, top_n=10, sat_range=50):
    user_school = data.loc[data['INSTNM'] == user_input]
    if user_school.empty:
        print("School not found!")
        return
    
    user_features = user_school[['LATITUDE', 'LONGITUDE', 'SAT_AVG']].iloc[0]
    data['similarity'] = data.apply(lambda x: euclidean_distance(user_features, x[['LATITUDE', 'LONGITUDE', 'SAT_AVG']], sat_range), axis=1)
    similar_schools = data.sort_values(by='similarity').head(top_n)
    return similar_schools[['UNITID', 'INSTNM', 'CITY', 'STABBR', 'SAT_AVG']]

# Step 3: User input and display similar schools
user_input = input("Enter the name of a school: ")
similar_schools = find_similar_schools(user_input, new_df)
if similar_schools is not None:
    print("Top 10 most similar schools to", user_input, ":")
    print(similar_schools)
```
